## Formater les sorties

On parle d'interpolation (des valeurs dans la chaîne de caractères).

Les instructions de formatage pour contrôler le nombre de chiffres avant, après la virgule, les zéros non significatifs, etc. utilise une syntaxe particulière dont la règle est la suivante 😨 :

```
{:[[fill]align][sign][#][0][width][grouping_option][.precision][type]}
```

Plus pratique, l'[anti-sèche](https://www.pythoncheatsheet.org/cheatsheet/string-formatting)

````python
# % opérateur d'interpolation
pi = 3.141
"La valeur de Pi : %.2f" % pi
````

````python
# .format() deuxième opérateur
"La première valeur {:.2f} et la deuxième valeur {:.2f}".format(pi, 1.414)
````

````python
# Les f-strings (chaîne de caractères commençant par f"")
f"La première valeur {pi:.2f}"
````

````python
print(f"La première valeur {pi:.2f}")
````

## La structure conditionnelle

````python
if a > 5:
    print('a est supérieure à 5')
    print('autres instructions, etc.')
else:
    print("a n'est pas supérieure à 5")
    print('etc.')
````

````python
if a > 5:
    print('a est supérieure à 5')
    print('autres instructions, etc.')
elif a < 0:
    print('a est négative')
    print('autres instructions, etc.')    
else:
    print("a n'est pas supérieure à 5 et n'est pas négative")
    print('etc.')
````

## La boucle for

Pour itérer sur une collection généralement, avec le mot_clé `in`.

````python
print("Pour mémoire : ", ma_liste)

for element in ma_liste:
    print(element)

print('fini')
````

````python
# range(début, fin exclue[, pas=1])
for element in range(0, 4):
    print(element)
````

````python
# range(début, fin exclue[, pas])
for element in range(0, 10, 2):
    print(element)
````

````python
# Parcours moins "pythonique" de la liste
for indice in range(0, len(ma_liste)):
    print(ma_liste[indice])
````

````python
# A comparer à cette version très "pythonique" / pythonesque
for element in ma_liste:
    print(element)
````

````python
for indice, element in enumerate(ma_liste):
    print(indice, '->', element)
````

### Les _list comprehensions_

````python
quelques_entiers = [1, 2, 3, 4]
````

````python
[entier * 2 for entier in quelques_entiers]
````

````python
double_entiers = [entier * 2 for entier in quelques_entiers]
double_entiers
````

````python
# On peut ajouter une condition pour filtrer les éléments sur lesquels le calcul est appliqué
[entier * 2 for entier in quelques_entiers if entier > 2]
````

## Les fonctions

Les fonctions permettent de regrouper un code Python que l'on exécute plusieurs fois, dans différents contextes, etc.

Elles commencent par une instruction `def`, suivi d'un nom, suivi d'une liste d'arguments, `:`, le corps du programme à réaliser et le mot-clé `return` avec le résultat à retourner, s'il y en a un.

````python
def somme(a, b):
    return a + b
````

````python
somme(2, 3)
````

````python
def affiche_somme(a, b):
    print('Le résultat est :', a + b)
    return
````

````python
affiche_somme(2, 3)
````

````python
resultat = affiche_somme(2, 3)
````

````python
resultat
````

````python
resultat = somme(2, 3)
resultat
````

````python
def somme(a, b=2):
    return a + b
````

````python
somme(2)
````

````python
somme(2, 3)
````

````python
def somme(a, b=2, c=3):
    return a + b + c
````

````python
somme(2)
````

````python
somme(2, c=5)
````

### Les docstrings

Chaîne de caractères de documentation.

````python
def somme(a, b=1):
  """Add two numbers.

  Plus d'explications...
  
  Params:
  a (int): first number to add
  b (int): second number to add (default=1)

  Returns:
  (int): sum of a and b
  """
  return a + b
````

````python
somme(5)
````

````python
help(somme)
````

````python
def produit(a, b):
    """Multiply two numbers.

    Plus d'explications...

    Params:
    a (int): first number to multiply
    b (int): second number to multiply

    Returns:
    (int): product of a and b
    
    Examples:
    >>> produit(2, 3)
    6
    >>> produit(2, -3)
    -6
    """
    return a * b
````
